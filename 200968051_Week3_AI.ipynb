```python
class Node():
    def __init__(self, state, parent, action):
        self.state = state
        self.parent = parent
        self.action = action
```

```python
#utilities

class StackFrontier():
    def __init__(self):
        self.frontier = []

    def add(self, node):
        self.frontier.append(node)

    def contains_state(self, state):
        return any(node.state == state for node in self.frontier)

    def empty(self):
        return len(self.frontier) == 0

    def remove(self):
        if self.empty():
            raise Exception("empty frontier")
        else:
            node = self.frontier[-1]
            self.frontier = self.frontier[:-1]
            return node

class QueueFrontier(StackFrontier):

    def remove(self):
        if self.empty():
            raise Exception("empty frontier")
        else:
            node = self.frontier[0]
            self.frontier = self.frontier[1:]
            return node
```

```python
import csv
import sys

# Maps names to a set of corresponding person_ids
names = {}

# Maps person_ids to a dictionary of: name, birth, movies (a set of movie_ids)
people = {}

# Maps movie_ids to a dictionary of: title, year, stars (a set of person_ids)
movies = {}
```

```python
# Section for Loading data
def load_data(directory):
    """
    Load data from CSV files into memory.
    """
    # Load people
    with open(f"{directory}/people.csv", encoding="utf-8") as f:
        reader = csv.DictReader(f)
        for row in reader:
            people[row["id"]] = {
                "name": row["name"],
                "birth": row["birth"],
                "movies": set()
            }
            if row["name"].lower() not in names:
                names[row["name"].lower()] = {row["id"]}
            else:
                names[row["name"].lower()].add(row["id"])

    # Load movies
    with open(f"{directory}/movies.csv", encoding="utf-8") as f:
        reader = csv.DictReader(f)
        for row in reader:
            movies[row["id"]] = {
                "title": row["title"],
                "year": row["year"],
                "stars": set()
            }

    # Load stars
    with open(f"{directory}/stars.csv", encoding="utf-8") as f:
        reader = csv.DictReader(f)
        for row in reader:
            try:
                people[row["person_id"]]["movies"].add(row["movie_id"])
                movies[row["movie_id"]]["stars"].add(row["person_id"])
            except KeyError:
                pass
```

```python
def bfs_path(source, target):
    """
    Implements the Breadth First Search Algirhtm to search for the path between 
    source and target.

    Returns the shortest list of (movie_id, person_id) pairs
    that connect the source to the target.

    If no possible path, returns None.
    """

    # TODO

    # handling edge cases
    if source == None:
      return None

    if source == target:
      return []

    #initializing the starting node
    start = Node(source, None, None)
    #initializing the frontier
    frontier  = QueueFrontier()
    #adding the start/root node to frontier
    frontier.add(start)
    #initializing explored set for the graph based BFS
    exploredset = set()

    while True:

      # If frontier is empty, path was not found and hence return None
      if frontier.empty():
        print('Frontier is Empty - No Connection/Path Between Actors was found!')
        print(len(exploredset), 'actors were explored with no solution/path found between the source and target!')
        return None
      
      # popping frontmost node
      frontnode = frontier.remove()
      # fronmost mode has been explored
      exploredset.add(frontnode.state)
      #All neighbors of frontmost node
      neighbors = neighbors_for_person(frontnode.state)

      # iterate over all neighbors
      for action, state in neighbors:

        # ignore actor is previously encountered. This is done to avoid redundant paths.
        if frontier.contains_state(state) or state in exploredset:
          continue
        
        # if the state is required target create a path and append all nodes to reach there 
        if state == target:
          path = []
          path.append((action, state))

          while frontnode.parent != None:
            path.append((frontnode.action, frontnode.state))
            frontnode = frontnode.parent

          path.reverse()

          return path

        #if state is not target create a new node and append to frontier.
        newnode = Node(state, frontnode, action)
        frontier.add(newnode)
```

```python
def dfs_path(source, target):
    """
    Implements the Bepth First Search Algirhtm to search for the path between 
    source and target.

    Returns the shortest list of (movie_id, person_id) pairs
    that connect the source to the target.

    If no possible path, returns None.
    """

    # TODO

    # handling edge cases
    if source == None:
      return None

    if source == target:
      return []

    #initializing the starting node
    start = Node(source, None, None)
    #initializing the frontier
    frontier  = StackFrontier()
    #adding the start/root node to frontier
    frontier.add(start)
    #initializing explored set for the graph based BFS
    exploredset = set()

    while True:

      # If frontier is empty, path was not found and hence return None
      if frontier.empty():
        print('Frontier is Empty - No Connection/Path Between Actors was found!')
        print(len(exploredset), 'actors were explored with no solution/path found between the source and target!')
        return None
      
      
      # popping frontmost node
      frontnode = frontier.remove()
      # fronmost mode has been explored
      exploredset.add(frontnode.state)
      #All neighbors of frontmost node
      neighbors = neighbors_for_person(frontnode.state)

      #iterate over all neighbors
      for action, state in neighbors:
        
        # ignore actor is previously encountered. This is done to avoid redundant paths.
        if frontier.contains_state(state) or state in exploredset:
          continue
            
        if not frontier.contains_state(state) and state not in exploredset:
          child = Node(state, frontnode, action)

          # If node is the goal, then we have a solution
          if child.state == target:
            path = []
            while child.parent is not None:
              path.append((child.action, child.state))
              child = child.parent
            path.reverse()
            return path

          # Otherwise add node to the frontier
          else:
            frontier.add(child)
```

```python
def person_id_for_name(name):
    """
    Returns the IMDB id for a person's name,
    resolving ambiguities as needed.
    """
    person_ids = list(names.get(name.lower(), set()))
    if len(person_ids) == 0:
        return None
    elif len(person_ids) > 1:
        print(f"Which '{name}'?")
        for person_id in person_ids:
            person = people[person_id]
            name = person["name"]
            birth = person["birth"]
            print(f"ID: {person_id}, Name: {name}, Birth: {birth}")
        try:
            person_id = input("Intended Person ID: ")
            if person_id in person_ids:
                return person_id
        except ValueError:
            pass
        return None
    else:
        return person_ids[0]
```

```python
def neighbors_for_person(person_id):
    """
    Returns (movie_id, person_id) pairs for people
    who starred with a given person.
    """
    movie_ids = people[person_id]["movies"]
    neighbors = set()
    for movie_id in movie_ids:
      for person_id in movies[movie_id]["stars"]:
        neighbors.add((movie_id, person_id))
    return neighbors
```
